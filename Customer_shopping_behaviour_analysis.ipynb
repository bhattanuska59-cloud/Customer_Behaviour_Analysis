```python
df["Review Rating"]=df["Review Rating"].fillna(df["Review Rating"].median())
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
## create a column age_group
labels=["Young Adult","Adult","Middle-aged","senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
print(df['frequency_of_purchases'].unique())
```

```python
## create purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
     'Weekly':7,
     'Monthly':30,
     'Quarterly':90,
     'Bi-Weekly':14,
     'Annually':365,
     'Every 3 months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype('Int64')
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```
